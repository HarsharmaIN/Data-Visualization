# Histogram

```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
from scipy import stats
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
# Magic line function for plotting graphs in the ipynb notebook
```

```python
dataset = randn(100)
```

```python
plt.hist(dataset) 
#'bins' are the number of vertical rectangles you can see in the histogram
# By default number of bins = 10
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color = 'indianred')
```

## Plotting dataset1 and dataset2 in the same plot

```python
plt.hist(dataset, color = 'indianred', alpha = 0.5, bins = 20)
plt.hist(dataset2, alpha = 0.5, bins = 20)

# 'alpha': for adjusting tranparency levels of histogram
```
